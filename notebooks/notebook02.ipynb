```python
# Input text
text =  "Apple is looking at buying U.K. startup $1 billion in London"

doc = nlp(text)

# Tokenization
print("Tokens:")
for token in doc:
    print(token.text)

# Part-of-Speech (POS)
print("\nPOS Tags:")
for token in doc:
    print(token.text, token.pos_)

    # Named Entity Recognition (NER)
print("\nEntities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
